```python
# 위치 B의 데이터 필터링
df_B = df[df['위치'] == 'B']

# 월별 집계
df_B['월'] = df_B['일자'].dt.to_period('M')
monthly_data = df_B.groupby('월').agg({
    '현재발전출력': 'mean',
    '수평일사량1': 'mean',
    '수평일사량2': 'mean',
    '경사일사량1': 'mean',
    '경사일사량2': 'mean',
    '외기온도1': 'mean',
    '외기온도2' : 'mean'
}).reset_index()

# 월 열을 datetime 형식으로 변환
monthly_data['월'] = monthly_data['월'].dt.to_timestamp()

# 시계열 시각화
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10), sharex=True)

# 발전 출력 시계열
sns.lineplot(x='월', y='현재발전출력', data=monthly_data, ax=axes[0, 0], marker='o')
axes[0, 0].set_title('현재 발전 출력 시계열')

# 수평일사량1 시계열
sns.lineplot(x='월', y='수평일사량1', data=monthly_data, ax=axes[0, 1], color='orange', marker='o')
axes[0, 1].set_title('수평일사량1 시계열')

# 수평일사량2 시계열
sns.lineplot(x='월', y='수평일사량2', data=monthly_data, ax=axes[0, 2], color='red', marker='o')
axes[0, 2].set_title('수평일사량2 시계열')

# 경사일사량1 시계열
sns.lineplot(x='월', y='경사일사량1', data=monthly_data, ax=axes[1, 0], color='green', marker='o')
axes[1, 0].set_title('경사일사량1 시계열')

# 경사일사량2 시계열
sns.lineplot(x='월', y='경사일사량2', data=monthly_data, ax=axes[1, 1], color='blue', marker='o')
axes[1, 1].set_title('경사일사량2 시계열')

# 외기온도1 시계열
sns.lineplot(x='월', y='외기온도1', data=monthly_data, ax=axes[1, 2], color='purple', marker='o')
axes[1, 2].set_title('외기온도1 시계열')

# 외기온도1 시계열
sns.lineplot(x='월', y='외기온도2', data=monthly_data, ax=axes[1, 2], color='purple', marker='o')
axes[1, 2].set_title('외기온도2 시계열')

# 레이아웃 조정
plt.tight_layout()
plt.show()

# 상관관계 분석
correlation_matrix = monthly_data[['현재발전출력', '수평일사량1', '수평일사량2', '경사일사량1', '경사일사량2', '외기온도1','외기온도2']].corr()
print(correlation_matrix)
```

```python
import statsmodels.api as sm

# 독립 변수와 종속 변수 정의
X = monthly_data[['수평일사량1', '수평일사량2', '경사일사량1', '경사일사량2', '외기온도1','외기온도2']]
y = monthly_data['현재발전출력']

# 상수항 추가
X = sm.add_constant(X)

# 회귀 모델 적합
model = sm.OLS(y, X).fit()

# 회귀 결과 출력
print(model.summary())

# 잔차 플롯
residuals = model.resid
fitted = model.fittedvalues

plt.figure(figsize=(10, 6))
sns.scatterplot(x=fitted, y=residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor, OLSInfluence
# Q-Q 플롯
plt.subplot(1, 2, 2)
sm.qqplot(residuals, line='s', ax=plt.gca())
plt.title('Q-Q Plot')
plt.tight_layout()
plt.show()

# 다중 공선성 체크 (VIF)
vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print("\nVariance Inflation Factors (VIF):")
print(vif_data)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 적용
pca = PCA(n_components=X_scaled.shape[1])  # 주성분 개수는 설명 변수 개수와 같음
X_pca = pca.fit_transform(X_scaled)

# 주성분 설명 비율 출력
print("Explained Variance Ratio of Each Principal Component:")
print(pca.explained_variance_ratio_)

# 주성분 데이터로 DataFrame 생성
X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

# 회귀 모델 적합
X_pca_df = sm.add_constant(X_pca_df)  # 상수항 추가
model_pca = sm.OLS(y, X_pca_df).fit()

# PCA 모델 요약 출력
print("\nPCA Model Summary:")
print(model_pca.summary())

# PCA 주성분 시각화 (첫 두 주성분을 사용하여 시각화)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x=X_pca_df['PC1'], y=X_pca_df['PC2'], c=y, cmap='viridis')
plt.title('PCA: 첫 두 주성분')
plt.xlabel('PC1')
plt.ylabel('PC2')

# 색상 막대 추가
cbar = plt.colorbar(scatter)
cbar.set_label('현재발전출력')

plt.show()
```

```python
# PCA 해도 다중공선성있다고해서 
# 라쏘,릿지로 해결해봄
from sklearn.linear_model import LassoCV, RidgeCV

# PCA 적용
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# 주성분 설명 비율 출력
print("Explained Variance Ratio of Each Principal Component:")
print(pca.explained_variance_ratio_)

# 주성분 데이터로 DataFrame 생성
X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

# Lasso 회귀 모델 적합
lasso = LassoCV(alphas=[0.1, 1.0, 10.0], cv=5, max_iter=10000)
lasso.fit(X_pca_df, y)

print("\nLasso 회귀 결과:")
print("Lasso 회귀 계수:")
print(lasso.coef_)
print("Lasso 회귀 최적 alpha:", lasso.alpha_)

# Ridge 회귀 모델 적합
ridge = RidgeCV(alphas=[0.1, 1.0, 10.0], cv=5)
ridge.fit(X_pca_df, y)

print("\nRidge 회귀 결과:")
print("Ridge 회귀 계수:")
print(ridge.coef_)
print("Ridge 회귀 최적 alpha:", ridge.alpha_)

# PCA 회귀 모델 적합
X_pca_df = sm.add_constant(X_pca_df)  # 상수항 추가
model_pca = sm.OLS(y, X_pca_df).fit()

# PCA 모델 요약 출력
print("\nPCA 회귀 모델 요약:")
print(model_pca.summary())

# PCA 주성분 시각화 (첫 두 주성분을 사용하여 시각화)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x=X_pca_df['PC1'], y=X_pca_df['PC2'], c=y, cmap='viridis')
plt.title('PCA: 첫 두 주성분')
plt.xlabel('PC1')
plt.ylabel('PC2')

# 색상 막대 추가
cbar = plt.colorbar(scatter)
cbar.set_label('현재발전출력')

plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# 예측값 생성
y_pred_pca = model_pca.predict(X_pca_df)
y_pred_lasso = lasso.predict(X_scaled)
y_pred_ridge = ridge.predict(X_scaled)

# 성능 평가
mse_pca = mean_squared_error(y, y_pred_pca)
mse_lasso = mean_squared_error(y, y_pred_lasso)
mse_ridge = mean_squared_error(y, y_pred_ridge)

r2_pca = r2_score(y, y_pred_pca)
r2_lasso = r2_score(y, y_pred_lasso)
r2_ridge = r2_score(y, y_pred_ridge)

print(f"PCA 회귀 MSE: {mse_pca:.2f}, R2: {r2_pca:.2f}")
print(f"Lasso 회귀 MSE: {mse_lasso:.2f}, R2: {r2_lasso:.2f}")
print(f"Ridge 회귀 MSE: {mse_ridge:.2f}, R2: {r2_ridge:.2f}")
```

```python
from sklearn.cluster import KMeans

# 독립 변수와 종속 변수 정의
X = df[['수평일사량1', '수평일사량2', '경사일사량1', '경사일사량2', '외기온도1', '외기온도2']]
y = df['현재발전출력']

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 적용
pca = PCA(n_components=2)  # 주성분 2개 선택
X_pca = pca.fit_transform(X_scaled)

# K-평균 클러스터링 적용
kmeans = KMeans(n_clusters=3, random_state=42)  # K 값을 3으로 설정
clusters = kmeans.fit_predict(X_pca)

# PCA 주성분 데이터프레임 생성
X_pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
X_pca_df['Cluster'] = clusters
X_pca_df['현재발전출력'] = y  # 종속 변수 추가

# 시각화
plt.figure(figsize=(12, 8))
scatter = plt.scatter(x=X_pca_df['PC1'], y=X_pca_df['PC2'], c=X_pca_df['Cluster'], cmap='viridis', edgecolor='k', alpha=0.7)
plt.colorbar(scatter, label='클러스터')
plt.title('PC1 vs PC2 with K-means Clustering')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()

# 클러스터 중심 시각화
plt.figure(figsize=(12, 8))
plt.scatter(x=X_pca_df['PC1'], y=X_pca_df['PC2'], c=X_pca_df['Cluster'], cmap='viridis', edgecolor='k', alpha=0.7)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='클러스터 중심')
plt.colorbar(label='클러스터')
plt.title('K-평균 클러스터링과 클러스터 중심을 사용한 PC1과 PC2')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.legend()
plt.show()
```

### 계절적 패턴 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 데이터 준비 (여기서는 예시 데이터프레임을 사용)
# monthly_data는 원본 데이터프레임입니다.
# monthly_data = pd.read_csv('data.csv')  # 예를 들어 데이터를 불러오는 코드

# 위치별 데이터 필터링
locations = df['위치'].unique()
for location in locations:
    df_location = df[df['위치'] == location]
    
    # 데이터 전처리
    X = df_location[['수평일사량1', '수평일사량2', '경사일사량1', '경사일사량2', '외기온도1', '외기온도2']]
    y = df_location['현재발전출력']

    # 데이터 표준화
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # PCA 적용
    pca = PCA(n_components=2)  # 첫 두 주성분만 사용
    X_pca = pca.fit_transform(X_scaled)

    # K-평균 클러스터링 적용
    kmeans = KMeans(n_clusters=3, random_state=42)  # 클러스터 수를 3으로 설정
    clusters = kmeans.fit_predict(X_pca)

    # 데이터프레임에 클러스터 레이블 추가
    df_location['Cluster'] = clusters

    # 클러스터별 평균 발전 출력 계산
    cluster_means = df_location.groupby('Cluster')['현재발전출력'].mean()
    
    # 시각화: 클러스터별 발전 출력 평균
    plt.figure(figsize=(12, 6))
    cluster_means.plot(kind='bar', color='skyblue')
    plt.title(f'{location} 위치의 클러스터별 발전 출력 평균')
    plt.xlabel('클러스터')
    plt.ylabel('현재 발전 출력')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# 데이터 준비: 예시로 데이터프레임을 사용합니다
# df = pd.read_csv('data.csv')  # 데이터를 불러오는 코드

# 데이터 전처리: 일사량과 외기온도, 현재발전출력 선택
X = df[['수평일사량1', '수평일사량2', '외기온도1', '외기온도2']]
y = df['현재발전출력']

# 숫자형 데이터 확인 및 변환
df['현재발전출력'] = pd.to_numeric(df['현재발전출력'], errors='coerce')

# 숫자가 아닌 값이 있는 행을 제거합니다.
df = df.dropna(subset=['현재발전출력'])

# 데이터 전처리
X = df[['수평일사량1', '수평일사량2', '외기온도1', '외기온도2']]
y = df['현재발전출력']

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 적용
pca = PCA(n_components=2)  # 첫 두 주성분만 사용
X_pca = pca.fit_transform(X_scaled)

# K-평균 클러스터링 적용
kmeans = KMeans(n_clusters=3, random_state=42)  # 클러스터 수를 3으로 설정
clusters = kmeans.fit_predict(X_pca)

# 클러스터 레이블 추가
df = df.copy()
df['Cluster'] = clusters

# 클러스터의 중심을 기준으로 계절 정의
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=['PC1', 'PC2'])
cluster_centers['Cluster'] = range(len(cluster_centers))

# 클러스터를 계절로 정의
def assign_season(cluster_center):
    if cluster_center['PC1'] > 0 and cluster_center['PC2'] > 0:
        return 0  # 여름
    elif cluster_center['PC1'] > 0 or cluster_center['PC2'] > 0:
        return 1  # 봄/가을
    else:
        return 2  # 겨울

cluster_centers['Season'] = cluster_centers.apply(assign_season, axis=1)

# 클러스터와 계절 레이블 결합
df = df.merge(cluster_centers[['Cluster', 'Season']], on='Cluster', suffixes=('', '_centers'))

# 위치별 데이터 필터링
locations = df['위치'].unique()
for location in locations:
    df_location = df[df['위치'] == location]
    
    # 클러스터별 현재 발전 출력의 평균
    cluster_means_region = df_location.groupby('Season')['현재발전출력'].mean()

    # 시각화: 각 지역별 클러스터링에 따른 발전 출력
    plt.figure(figsize=(10, 6))
    cluster_means_region.plot(kind='bar', color=['r', 'g', 'b'])
    plt.title(f'{location} 위치의 클러스터별 현재 발전 출력')
    plt.xlabel('계절')
    plt.ylabel('현재 발전 출력')
    plt.xticks(rotation=0)
    plt.grid(True)
    plt.show()
```
